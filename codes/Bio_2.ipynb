```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
records=[]

for record in SeqIO.parse("gene.fna","fasta"):
  rc=record.seq.reverse_complement()
  records.append(SeqRecord(rc,id=record.id))
  SeqIO.write(records,"reverse.fasta","fasta")

```

```python
from Bio import SeqIO

for record in SeqIO.parse("gene.fna","fasta"):
  seq=str(record.seq)
  for i in range(len(seq)):
    if seq[i:i+3]=="ATG":
      print("ATG at",i)
```

```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord

proteins=[]
for record in SeqIO.parse("gene.fna","fasta"):
  proteins.append(SeqRecord(record.seq.translate(),id=record.id))

  SeqIO.write(proteins,"proteins.fasta","fasta")
  print(record.seq)
```
